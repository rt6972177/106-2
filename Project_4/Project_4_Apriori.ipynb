```python
import pandas as pd
import pandas as pd
from mlxtend.frequent_patterns import apriori
from mlxtend.frequent_patterns import association_rules
```

```python
df_orders = pd.read_csv('91APP_Orders_promotions.csv')
df_promo = pd.read_csv('91APP_promotion.csv')
df_promo_orders = pd.read_csv('91APP_PromotionOrders.csv')
print(df_orders.columns)
print('\n')
print(df_promo.columns)
print('\n')
print(df_promo_orders.columns)
```

```python
df_orders_need = df_orders[['TradesOrderGroupCode','LevelFourCategoryName','SalesOrderSlaveId']]
```

```python
print('Number of orders is',str(df_orders_need['SalesOrderSlaveId'].nunique()))
```

```python
df_new = df_orders_need[['TradesOrderGroupCode','LevelFourCategoryName']]
```

### 製作Apriori需要的Table

```python
df_new['quantity'] = 1
df_new
```

```python
df_new_sum = df_new.groupby(by=['TradesOrderGroupCode','LevelFourCategoryName'])['quantity'].sum()
```

```python
df_new_one_hot = df_new_sum.unstack().fillna(0).applymap(lambda x: 1 if x>0 else 0)
df_new_one_hot_reTshit = df_new_one_hot.drop('T恤/帽T',axis=1)
df_new_one_hot_reTshit
```

### 計算每個品項購買的支持度，最高的是T-shirt，有接近一半的機率(留待後用)

```python
frequent_itemsets = apriori(df_new_one_hot, min_support=0.1, use_colnames=True)
frequent_itemsets.sort_values(by = 'support',ascending=False)
```

```python
frequent_itemsets = apriori(df_new_one_hot, min_support=0.01, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.5)
rules.sort_values(by = 'confidence',ascending=False)
```

### 發現信度的後項幾乎都被T-shirt佔據了，但T-shirt購買的機會較高蠻符合直覺，所以刪去T-shirt的資料再計算一次

```python
frequent_itemsets = apriori(df_new_one_hot_reTshit, min_support=0.01, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.35)
rules.sort_values(by = 'confidence',ascending=False)
```

### 發現購買高領衫、毛衣和圍巾的人，會購買針織衫的信度都比較高，推論天氣冷的時候這些類型的厚衣服的搭配都是必須的

```python
df_new_one_hot_reTshit_dress = df_new_one_hot_reTshit.drop('洋裝',axis=1)
```

```python
frequent_itemsets = apriori(df_new_one_hot_reTshit_dress, min_support=0.01, use_colnames=True)
rules = association_rules(frequent_itemsets, metric="confidence", min_threshold=0.3)
rules.sort_values(by = 'confidence',ascending=False)
```
